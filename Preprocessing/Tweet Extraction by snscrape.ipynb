```python
import pandas as pd
df= pd.read_csv("merged_data.csv")
```

```python
df.head()
```

```python
df.dropna(subset = ['tweet_ids'], axis = 0, inplace=True)
```

```python
import re
from re import split

df['tweet_ids'] = df['tweet_ids'].astype(str)

df['tweet_ids'] = df['tweet_ids'].apply(lambda x: str(x).replace('\t',' '))
```

```python
df
```

```python
df['tweet_ids'] = df['tweet_ids'].apply(lambda x: re.split(r' ', x))
df
```

```python
df['number_tweets']=df['tweet_ids'].apply(lambda x: len(x)).astype(int)
df
```

```python
df['id'].value_counts()
```

```python
df.reset_index(inplace = True)
```

```python
from tqdm import tqdm
import snscrape.modules.twitter as sntwitter

tweet_data = []
for x in tqdm(range(len(df))):
    for t_id in df['tweet_ids'][x]:
        for i,tweet in enumerate(sntwitter.TwitterTweetScraper(tweetId=t_id,mode=sntwitter.TwitterTweetScraperMode.SINGLE).get_items()):
            tweet_data.append([df['id'][x],tweet.id, tweet.date, tweet.content,tweet.retweetCount,tweet.likeCount,tweet.replyCount,tweet.quoteCount])

tweet_df = pd.DataFrame(tweet_data, columns=['Id','Tweet_Id','Datetime','Text','Retweet','Like','Reply','Quote'])
```
